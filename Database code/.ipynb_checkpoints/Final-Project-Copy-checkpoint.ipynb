```python
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso
import numpy as np
#import seaborn as sns
#import scipy as sp
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
import pickle
```

```python
data = pd.read_csv('full_data.csv')
```

```python
data = data.drop('Europe', axis=1)
```

```python
data = data.dropna()
```

### XGBoost Model

```python
from numpy import loadtxt
import xgboost as xgb
from xgboost import XGBRegressor,plot_importance
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import cross_val_score,KFold
#from sklearn.grid_search import GridSearchCV   #Perforing grid search
from scipy.stats import skew
from collections import OrderedDict
```

```python
X = data[['High_School', 'Bachelors', 'Never_Married', 'Married',
       'Male', 'Black', 'Hispanic_Latino', 'White', 'Unemployment', '9th_12th']]
Y = data["Violent_Crime"]
```

```python
X= X.to_numpy()
Y= Y.to_numpy()
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.33, random_state=42)
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = .70, learning_rate = 0.2,
                max_depth = 3, alpha = 50, n_estimators =500, min_child_weight=1, colsample_bylevel=.70)
```

```python
xg = xg_reg.fit(X_train, y_train)
preds = xg_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(xg_reg, X_test, y_test, cv=5)
scores
```

```python
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
